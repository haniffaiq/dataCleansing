```python
data = pd.read_csv('TempDataAktifitas.csv')
data = pd.DataFrame(data)

print(data.columns)
```

```python
def isnan(value):
    try:
        import math
        return math.isnan(float(value))
    except:
        return False
```

```python
kolom = ['No.', 'Tanggal', 'Nomor', 'ATM ID', 'Lokasi', 'Jenis Layanan','Laporan Kerusakan', 'Pekerjaan Dilakukan', 'Nomor Tiket','Status', 'CSE', 'Pilihan']
for i in range (len(data)):
    
    #print(type(data.iloc[i]['Tanggal']))
    x = data.iloc[i]['Tanggal']
    if(isnan(x)):
        idx = i
        beforeNaN = i - 1
        for j in range(len(kolom)):
            
            data.at[idx, kolom[j]] = data.iloc[beforeNaN][kolom[j]]
    
  
```
